```python
import matplotlib.pyplot as plt
import numpy as np

alpha=0.3

u1 = lambda x: 1*x
u2 = lambda x: 5*x

f1 = lambda x: u1(x)**alpha
f2 = lambda x: u2(x)**alpha

x = np.array([i for i in range(1,11)])

plt.figure()
plt.plot(x, np.ones(10)*1, '-o', color='blue', label='country 1')
plt.plot(x, np.ones(10)*5, '-o', color='red', label='country 2')
plt.title('GHG')
plt.legend()
plt.grid()
plt.show()

plt.figure()
plt.plot(x, u1(x), '-o', color='blue', label='u1')
plt.plot(x, u2(x), '-o', color='red', label='u2')
plt.title('Sum')
plt.legend()
plt.grid()
plt.show()

plt.figure()
plt.plot(x, f1(x), '--o', color='blue', label='f1')
plt.plot(x, f2(x), '--o', color='red', label='f2')
plt.title('Sum^alpha')
plt.legend()
plt.grid()
plt.show()
```
